```python
import pandas as pd

from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
transactions = [
    ['Milk', 'Bread', 'Butter'],
    ['Bread', 'Butter'],
    ['Milk', 'Bread'],
    ['Milk', 'Eggs'],
    ['Bread', 'Eggs'],
    ['Milk', 'Bread', 'Eggs', 'Butter'],
    ['Milk', 'Butter'],
    ['Bread', 'Butter', 'Eggs'],
    ['Milk', 'Bread', 'Butter', 'Eggs'],
    ['Milk', 'Bread'],
    ['Rice', 'Oil'],
    ['Rice', 'Oil', 'Sugar'],
    ['Milk', 'Sugar'],
    ['Bread', 'Jam'],
    ['Milk', 'Bread', 'Jam'],
    ['Butter', 'Jam'],
    ['Milk', 'Bread', 'Butter'],
    ['Eggs', 'Bread'],
    ['Milk', 'Rice'],
    ['Milk', 'Bread', 'Oil']
]

print("Transaction Dataset")
for i, t in enumerate(transactions, 1):
    print(i, ":", t)
```

```python
te = TransactionEncoder()

te_array = te.fit(transactions).transform(transactions)

df = pd.DataFrame(te_array, columns=te.columns_)

print("\nPreprocessed Dataset")
print(df.head())
```

```python
frequent_itemsets = apriori(
    df,
    min_support=0.20,
    use_colnames=True
)

print("\nFrequent Itemsets")
print(frequent_itemsets)
```

```python
rules = association_rules(
    frequent_itemsets,
    metric="confidence",
    min_threshold=0.60
)

print("\nAssociation Rules")
print(rules[['antecedents',
             'consequents',
             'support',
             'confidence',
             'lift']])
```

```python
print("\nRecommendation Insights\n")

for index, row in rules.iterrows():

    left = ", ".join(list(row['antecedents']))
    right = ", ".join(list(row['consequents']))

    print(f"If a customer buys [{left}]")
    print(f"Recommend: [{right}]")
    print("Confidence :", round(row['confidence'],2))
    print("Lift       :", round(row['lift'],2))
    print("-----------------------------------")
```

```python
frequent_itemsets.to_csv(
    "Frequent_Itemsets.csv",
    index=False
)

rules.to_csv(
    "Association_Rules.csv",
    index=False
)

print("\nAnalysis Completed Successfully!")
print("Frequent itemsets saved as Frequent_Itemsets.csv")
print("Association rules saved as Association_Rules.csv")
```
