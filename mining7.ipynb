```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("heartmining.csv")

# Display first rows
print("First 5 Records:")
print(df.head())

print("\nDataset Shape:", df.shape)
```

```python
print("\nDataset Information")
print(df.info())

print("\nMissing Values")
print(df.isnull().sum())
```

```python
df = df.drop_duplicates()

# Fill missing numerical values (if any)
for col in df.select_dtypes(include=np.number).columns:
    df[col].fillna(df[col].median(), inplace=True)

print("\nAfter Preprocessing")
print(df.isnull().sum())

# Statistical Summary
print("\nStatistical Summary")
print(df.describe())
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='target', data=df)
plt.title("Heart Disease Distribution")
plt.xlabel("Heart Disease (0=No, 1=Yes)")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(7,4))
sns.histplot(df['age'], bins=20, kde=True)
plt.title("Age Distribution")
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title("Feature Correlation")
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.boxplot(x='target', y='age', data=df)
plt.title("Age vs Heart Disease")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='sex', hue='target', data=df)
plt.title("Gender vs Heart Disease")
plt.xlabel("Gender (0=Female,1=Male)")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='cp', hue='target', data=df)
plt.title("Chest Pain Type vs Disease")
plt.show()
```

```python
print("\n================ MEDICAL INSIGHTS REPORT ================\n")

total_patients = len(df)
disease = df['target'].sum()
healthy = total_patients - disease

print("Total Patients :", total_patients)
print("Patients with Heart Disease :", disease)
print("Healthy Patients :", healthy)

print("\nAverage Age :", round(df['age'].mean(),2))

male = len(df[df['sex']==1])
female = len(df[df['sex']==0])

print("Male Patients :", male)
print("Female Patients :", female)

print("\nAverage Cholesterol :", round(df['chol'].mean(),2))
print("Average Blood Pressure :", round(df['trestbps'].mean(),2))
print("Average Maximum Heart Rate :", round(df['thalach'].mean(),2))

print("\nDisease Percentage :",
      round((disease/total_patients)*100,2),"%")

print("\n=========================================================")
```
